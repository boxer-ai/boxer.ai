```python
import nupic
import json
from collections import defaultdict

reviews_by_biz = defaultdict(list)
reviews_by_user = defaultdict(list)
bizs = {}
users = {}
  
with open('yelp_academic_dataset.json') as f:
    for l in f:
        parsed = json.loads(l)
        type = parsed['type']
        if type == 'user':
            users[parsed['user_id']] = parsed
        elif type == 'business':
            bizs[parsed['business_id']] = parsed
        elif type == 'review':
            reviews_by_biz[parsed['business_id']].append(parsed)
            reviews_by_user[parsed['user_id']].append(parsed)
        else:
            print parsed
    
    
```

```python
from pprint import pprint as pp
sample_bizs = bizs.values()[:500]
pp( [(b['name'], b["business_id"], b["review_count"]) for b in sample_bizs if b['review_count'] == 10])
```

```python
from pycept import Cept
from pprint import pprint as pp

cortical_key = open('cortical_io.key').read()

client = Cept(cortical_key, retina='eng_gen')
```

```python
from operator import itemgetter

def biz_to_sdr(cortical_client, biz_id, exclude_user=None):
    all_reviews = [r['text'] for r in reviews_by_biz[biz_id] if not r['user_id'] == exclude_user]
    all_review_text = ('. ').join(all_reviews)
    return cortical_client.getBitmap(all_review_text)
    
def biz_to_keywords(cortical_client, biz_id, exclude_user=None):
    biz_sdr = biz_to_sdr(cortical_client, biz_id, exclude_user=exclude_user)
    keywords = cortical_client.bitmapToTerms(biz_sdr['positions'])
    return (sorted([(w['term'], w['score']) for w in keywords], key=itemgetter(1), reverse=True))

def pp_biz(biz_id):
    for r in reviews_by_biz[biz_id]:
        print "user: %s  (%s)\n" % (users[r['user_id']]['name'], r['user_id'])
        print r["text"]
        print "\n------"
        
def pp_review(user_id, biz_id):
    user_biz_reviews = [r for r in reviews_by_user[user_id] if r['business_id'] == biz_id]
    for r in user_biz_reviews:
        print "%s writes... (%s)\n" % (users[r['user_id']]['name'], r['user_id'])
        print r["text"]
        print "\n------"

def review_to_sdr(cortical_client, user_id, biz_id, as_keywords=False):
    user_biz_reviews = [r['text'] for r in reviews_by_user[user_id] if r['business_id'] == biz_id]
    review_text = ('. ').join(user_biz_reviews)
    print review_text
    bmp = cortical_client.getBitmap(review_text)
    if as_keywords:
        return cortical_client.bitmapToTerms(bmp['positions'])
    return bmp

def user_contribution(cortical_client, user_id, biz_id, as_keywords=False):
    biz_sdr = biz_to_sdr(cortical_client, biz_id, exclude_user=user_id)
    user_review_sdr = review_to_sdr(cortical_client, user_id, biz_id)
    diff = cortical_client.bitmapSubBitmap(user_review_sdr['positions'], biz_sdr['positions'])
    if as_keywords:
        return cortical_client.bitmapToTerms(diff)
    return diff


    
    
    
```
